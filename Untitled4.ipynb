```python
# ================================================
# 🔍 DÉTECTEUR DE FRAUDES BANCAIRES
# Auteur : God Win A. Real FADONOUGBO
# Date   : 18 Avril 2026
# Stack  : Python · PyTorch · Scikit-learn
# ================================================

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd

# ================================
# 📦 GÉNÉRATION DES DONNÉES
# ================================
X, y = make_classification(
    n_samples    = 1000,   # 1000 transactions
    n_features   = 10,     # 10 caractéristiques par transaction
    n_informative= 6,      # dont 6 vraiment utiles
    n_classes    = 2,      # 2 classes : fraude / légitime
    weights      = [0.95, 0.05],  # 5% de fraudes (réaliste !)
    random_state = 42
)

# Aperçu des données
df = pd.DataFrame(X, columns=[f"feature_{i}" for i in range(10)])
df['Class'] = y
print("Shape :", df.shape)
print("\nDistribution des classes :")
print(df['Class'].value_counts())
print(f"\nPourcentage de fraudes : {df['Class'].mean():.2%}")

#Séparation des données
X = df.drop('Class', axis=1).values
y=df['Class'].values

#Normalisation des données
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)

#Séparation des données
split = int(0.8 * len(X))          
X_train, X_test = X_scaled[:split], X_scaled[split:]
y_train, y_test = y[:split], y[split:]

#Transformation des données en tenseurs pytorch
X_train_py=torch.FloatTensor(X_train)
X_test_py=torch.FloatTensor(X_test)
y_train_py=torch.FloatTensor(y_train).unsqueeze(1)
y_test_py=torch.FloatTensor(y_test).unsqueeze(1)

#Début
train_loader=DataLoader(TensorDataset(X_train_py,y_train_py),batch_size=32,shuffle=True)
test_loader=DataLoader(TensorDataset(X_test_py,y_test_py),batch_size=32,shuffle=True)
print(f"X_train : {X_train_py.shape}")
print(f"X_test  : {X_test_py.shape}")
print(f"y_train : {y_train_py.shape}")
print(f"y_test  : {y_test_py.shape}")

# ============================================
# 2️⃣ MODÈLE, LOSS, OPTIMISEUR
# ============================================

modele=nn.Sequential(
    nn.Linear(10,32),
    nn.ReLU(),
    nn.Linear(32,16),
    nn.ReLU(),
    nn.Linear(16,1),
    nn.Sigmoid())
print(modele)
total_params=sum(p.numel() for p in modele.parameters())
print("Paramètres totaux:",total_params)

loss_fn=nn.BCELoss()
optimiseur=torch.optim.Adam(modele.parameters(),lr=0.001)


# ============================================
# 2️⃣ BOUCLE D'ENTRAINEMENT
# ============================================
def entrainer(modele,loader,loss_fn,optimiseur):
    modele.train()          # 
    total_loss = 0

    for X_batch, y_batch in loader:

        # Cycle complet
        optimiseur.zero_grad()              
        predictions = modele(X_batch)       
        loss = loss_fn(predictions, y_batch)
        loss.backward()                     
        optimiseur.step()                   

        total_loss += loss.item()

    return total_loss / len(loader)         


# ============================================
# 2️⃣ BOUCLE D'EVALUATION
# ============================================
def evaluer(modele,loader,loss_fn):
    modele.eval()           
    total_loss = 0
    correct    = 0

    with torch.no_grad():   #
        for X_batch, y_batch in loader:

            predictions = modele(X_batch)
            loss        = loss_fn(predictions, y_batch)
            total_loss += loss.item()

            pred_classes = (predictions > 0.5).float()
            correct += (pred_classes == y_batch).sum().item()

    accuracy = correct / len(loader.dataset)
    return total_loss / len(loader), accuracy


# ============================================
# 2️⃣ ENTRAINEMENT COMPLET
# ============================================



for epoch in range(50):
    train_loss             = entrainer(modele, train_loader, loss_fn, optimiseur)
    test_loss, test_acc    = evaluer(modele, test_loader, loss_fn)

    if epoch % 10 == 0:
        print(f"Epoch {epoch:3d} | "
              f"Train Loss: {train_loss:.4f} | "
              f"Test Loss: {test_loss:.4f} | "
              f"Accuracy: {test_acc:.2%}")
```
