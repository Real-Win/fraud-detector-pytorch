```python
#God Win FADONOUGBO
#03/05/2026
#20:36
import torch
import torch.nn as nn
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# Transformation : image → tenseur normalisé
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Téléchargement automatique !
train_data = datasets.MNIST(
    root='./data',
    train=True,
    download=True,
    transform=transform
)

test_data = datasets.MNIST(
    root='./data',
    train=False,
    download=True,
    transform=transform
)

# DataLoaders
train_loader = DataLoader(train_data,
                          batch_size=64,
                          shuffle=True)
test_loader  = DataLoader(test_data,
                          batch_size=64,
                          shuffle=False)

print(f"Train : {len(train_data)} images")
print(f"Test  : {len(test_data)} images")

# Vérification shape
images, labels = next(iter(train_loader))
print(f"Shape batch : {images.shape}")
print(f"Classes     : {train_data.classes}")
```

```python
# Afficher 10 images du dataset
fig, axes = plt.subplots(2, 5, figsize=(12, 5))

for i, ax in enumerate(axes.flat):
    ax.imshow(images[i].squeeze(), cmap='gray')
    ax.set_title(f"Label: {labels[i]}")
    ax.axis('off')

plt.suptitle("Exemples MNIST", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN,self).__init__()
        self.conv=nn.Sequential(
            nn.Conv2d(1,32,kernel_size=3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),
            nn.Conv2d(32,64,kernel_size=3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2))
        
        self.fc=nn.Sequential(
             nn.Linear(64*7*7,128), #pour le flatten
             nn.ReLU(),
             nn.Linear(128,64),
             nn.ReLU(),
             nn.Linear(64,10))
        
    def forward(self, x):
        x = self.conv(x)
        x = x.view(x.size(0), -1)  # Flatten
        return self.fc(x)

modele = CNN()
print(modele)
total = sum(p.numel() for p in modele.parameters())
print(f"Paramètres : {total:,}")
```

```python
loss_fn    = nn.CrossEntropyLoss()
optimiseur = torch.optim.Adam(modele.parameters(), lr=0.001)

def entrainer(modele, loader, loss_fn, optimiseur):
    modele.train()
    total_loss = 0
    correct    = 0

    for images, labels in loader:
        optimiseur.zero_grad()
        outputs = modele(images)              # forward pass
        loss    = loss_fn(outputs, labels)    # calcul loss
        loss.backward()                        # backward
        optimiseur.step()                      # update

        total_loss += loss.item()
        # Prédiction = classe avec le score le plus élevé
        preds    = outputs.argmax(dim=1)       # ← argmax pour multiclasse !
        correct += (preds == labels).sum().item()

    accuracy = correct / len(loader.dataset)
    return total_loss / len(loader), accuracy

def evaluer(modele,loader,loss_fn):
    modele.eval()
    total_loss=0
    correct =0
    with torch.no_grad():
        for images,labels  in loader:
            outputs=modele(images)
            loss=loss_fn(outputs,labels)
            total_loss += loss.item()
            # Prédiction = classe avec le score le plus élevé
            preds    = outputs.argmax(dim=1)       # ← argmax pour multiclasse !
            correct += (preds == labels).sum().item()
    accuracy=correct / len(loader.dataset)
    return total_loss / len(loader), accuracy


for epoch in range(5):
    loss_train, acc_train = entrainer(modele,train_loader,loss_fn,optimiseur)
    loss_test, acc_test = evaluer(modele,test_loader,loss_fn)

    print(f"Epoch {epoch+1}|Train_loss: {loss_train:.4f}|Test_loss: {loss_test:.4f}|Accuracy: {acc_test:.2%}")


    
    
```

```python
modele.eval()
images_test, labels_test = next(iter(test_loader))

#UN PETIT TEST POUR VERIFIER LA QUALITE DU MODELE
with torch.no_grad():
    outputs = modele(images_test[:10])
    preds   = outputs.argmax(dim=1)

fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for i, ax in enumerate(axes.flat):
    ax.imshow(images_test[i].squeeze(), cmap='gray')
    couleur = 'green' if preds[i] == labels_test[i] else 'red'
    ax.set_title(f"Prédit: {preds[i]}\nRéel: {labels_test[i]}",
                 color=couleur)
    ax.axis('off')

plt.suptitle("Prédictions CNN sur MNIST 🧠", fontsize=16)
plt.tight_layout()
plt.show()
```
