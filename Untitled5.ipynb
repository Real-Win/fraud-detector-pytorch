```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Téléchargement automatique via yfinance
# D'abord installe : pip install yfinance
import yfinance as yf

# Télécharger les prix Bitcoin
btc = yf.download('BTC-USD',
                   start='2020-01-01',
                   end='2024-01-01')

print(btc.head())
print(f"\nShape : {btc.shape}")
print(f"Colonnes : {btc.columns.tolist()}")

# Visualiser les prix
plt.figure(figsize=(12, 5))
plt.plot(btc['Close'])
plt.title('Prix Bitcoin 2020-2024 💰')
plt.xlabel('Date')
plt.ylabel('Prix (USD)')
plt.grid(True)
plt.show()
```

```python
# On utilise seulement le prix de clôture
prix = btc['Close'].values.reshape(-1, 1)

# Normalisation entre 0 et 1
scaler = MinMaxScaler()
prix_scaled = scaler.fit_transform(prix)

# Créer des séquences
# → On utilise 60 jours pour prédire le jour suivant
def creer_sequences(data, seq_length=60):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])      # 60 jours
        y.append(data[i+seq_length])         # jour suivant
    return np.array(X), np.array(y)

seq_length = 60
X, y = creer_sequences(prix_scaled, seq_length)

print(f"X shape : {X.shape}")
print(f"y shape : {y.shape}")

# Split train/test (80/20)
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Convertir en tenseurs
X_train = torch.FloatTensor(X_train)
X_test  = torch.FloatTensor(X_test)
y_train = torch.FloatTensor(y_train)
y_test  = torch.FloatTensor(y_test)

print(f"\nX_train : {X_train.shape}")
print(f"X_test  : {X_test.shape}")
print(f"y_train : {y_train.shape}")
print(f"y_test  : {y_test.shape}")
```

```python
class LSTMBitcoin(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super().__init__()

        self.hidden_size = hidden_size
        self.num_layers  = num_layers

        # Couche LSTM
        self.lstm = nn.LSTM(
            input_size  = input_size,   # 1 feature
            hidden_size = hidden_size,  # 64 neurones
            num_layers  = num_layers,   # 2 couches
            batch_first = True          # (batch, seq, features)
        )

        # Couche de sortie → prédit 1 prix
        self.fc = nn.Linear(hidden_size, 1)
        # ⚠️ Pas de Sigmoid → régression !

    def forward(self, x):
        # États cachés initialisés à zéro
        h0 = torch.zeros(self.num_layers,
                         x.size(0),
                         self.hidden_size)
        c0 = torch.zeros(self.num_layers,
                         x.size(0),
                         self.hidden_size)

        # Passage dans LSTM
        out, _ = self.lstm(x, (h0, c0))
        # out shape : (batch, seq_length, hidden_size)

        # Dernier pas de temps uniquement
        out = out[:, -1, :]
        # shape : (batch, hidden_size)

        # Prédiction finale
        out = self.fc(out)
        # shape : (batch, 1) → prix prédit !

        return out

modele = LSTMBitcoin(
    input_size  = 1,
    hidden_size = 64,
    num_layers  = 2
)

print(modele)
total = sum(p.numel() for p in modele.parameters())
print(f"Paramètres : {total:,}")
```

```python
from torch.utils.data import TensorDataset, DataLoader

# DataLoaders
train_loader = DataLoader(
    TensorDataset(X_train, y_train),
    batch_size=32, shuffle=True
)
test_loader = DataLoader(
    TensorDataset(X_test, y_test),
    batch_size=32, shuffle=False
)

# Fonctions
def entrainer(modele, loader, loss_fn, optimiseur):
    modele.train()
    total_loss = 0

    for X_batch, y_batch in loader:
        optimiseur.zero_grad()
        preds = modele(X_batch)
        loss  = loss_fn(preds, y_batch)
        loss.backward()
        optimiseur.step()
        total_loss += loss.item()

    return total_loss / len(loader)  # ← juste la loss !

def evaluer(modele, loader, loss_fn):
    modele.eval()
    total_loss = 0

    with torch.no_grad():
        for X_batch, y_batch in loader:
            preds      = modele(X_batch)
            loss       = loss_fn(preds, y_batch)
            total_loss += loss.item()  # ← important !

    return total_loss / len(loader)

# Boucle d'entraînement
for epoch in range(50):
    loss_train = entrainer(modele, train_loader,
                           loss_fn, optimiseur)
    loss_test  = evaluer(modele, test_loader, loss_fn)

    if epoch % 10 == 0:
        print(f"Epoch {epoch+1:3d} | "
              f"Train Loss: {loss_train:.6f} | "
              f"Test Loss:  {loss_test:.6f}")
```

```python
# Prédictions vs Réalité
modele.eval()
with torch.no_grad():
    preds_train = modele(X_train).numpy()
    preds_test  = modele(X_test).numpy()

# Dénormaliser les prix
preds_train = scaler.inverse_transform(preds_train)
preds_test  = scaler.inverse_transform(preds_test)
y_train_reel = scaler.inverse_transform(y_train.numpy())
y_test_reel  = scaler.inverse_transform(y_test.numpy())

# Visualisation
plt.figure(figsize=(14, 6))

# Prix réels
plt.plot(y_train_reel, label='Réel Train',
         color='blue', linewidth=1)
plt.plot(range(len(y_train_reel),
               len(y_train_reel) + len(y_test_reel)),
         y_test_reel, label='Réel Test',
         color='green', linewidth=1)

# Prédictions
plt.plot(preds_train, label='Prédit Train',
         color='orange', linewidth=1,
         linestyle='--')
plt.plot(range(len(preds_train),
               len(preds_train) + len(preds_test)),
         preds_test, label='Prédit Test',
         color='red', linewidth=1,
         linestyle='--')

plt.title('LSTM — Prédiction Prix Bitcoin 💰🔥')
plt.xlabel('Jours')
plt.ylabel('Prix (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
